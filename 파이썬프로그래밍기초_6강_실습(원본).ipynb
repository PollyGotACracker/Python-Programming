**5회 반복 출력**

```python
# 메시지 저장
msg = "저는 파이썬을 잘 합니다."

# 5회 반복 출력
i = 1
while i <= 5 :
  print(msg)
  i = i + 1
```

**i부터 n까지의 합(3단 전략)**

```python
i = 1
sum = 0
n = int(input("합산할 숫자를 입력하세요: "))
while i <= n :
  sum = sum + i
  i += 1

print("1부터", str(n)+"까지의 합은", str(sum)+"입니다.")
```

**사용자가 입력한 단의 구구단 출력 문제**

```python
base = int(input("출력할 단을 입력하세요: "))

i = 1
while i <= 9 :
  print(base, "x", i, "=", base * i)
  i += 1
```

# 6강 2절

**반지름은 10이고 높이가 1, 5, 14, 26, 31인 원뿔의 부피와 겉넓이를 각각 출력하는 문제**

```python
rad = 10
# 리스트 생성
hei_list = [1, 5, 14, 26, 31]

for hei in hei_list :
  # 모선 계산
  sla = pow(rad ** 2 + hei ** 2, 0.5)
  # 부피 & 겉넓이 출력
  vol = 1 / 3 * 3.14 * rad ** 2 * hei
  suf = 3.14 * rad ** 2 + 3.14 * rad * sla
  print("반지름:", rad, "높이:", hei, "원뿔")
  print("부피:", vol, "겉넓이:", suf)
```

**유클리드 호제법**

```python
# 유클리드 호제법으로 두 수의 최대공약수(GCD)를 구하는 프로그램

# 두 수 입력 받기
a = int(input("첫 번째 숫자를 입력하세요: "))
b = int(input("두 번째 숫자를 입력하세요: "))

# a를 b로 나눈 나머지를 b에 저장
# b가 0이 될 때까지 반복(b가 0이 아니라면 계속 반복)

while b != 0 :
  rem = a % b
  a = b
  b = rem
# 나머지가 0이 아닐 때 제수가 b이고, b를 a에 할당하였으므로 최대공약수는 a
print("두 수의 최대공약수는", str(a) + "입니다.")

# 반복구조 설계 전략
# 1. 반복되어야 할 명령블록 작성
# 2. 명령 블록을 while 안에 넣기
# 3. 반복-계속-조건 작성, 반복제어명령문 추가
```
