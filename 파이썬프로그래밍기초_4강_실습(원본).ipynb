## 원뿔의 부피&겉넓이 계산 프로그램 5


```python
# input(): 받은 값은 문자열로 처리
rad = input()
# type(): 어떤 형태의 데이터인지 반환하는 함수
print(type(rad))

rad = int(rad)
# 동적 타이핑 언어: 입력되는 값에 따라 변수의 타입이 변경
print(type(rad))
```

```python
# 원뿔 부피 및 겉넓이 계산 프로그램
rad = int(input("반지름을 입력하세요: "))
hei = int(input("높이를 입력하세요: "))
sla = pow(rad ** 2 + hei ** 2, 0.5)

# 부피 & 겉넓이 출력
# print()에서 여러 데이터를 한번에 출력 가능
  # comma로 연결: 구분자(sep)의 기본값은 공백
  # + 연산자로 연결: 구분자 없이 연결 가능. 단, 같은 데이터 타입이어야 함
  # str()을 사용하여 문자열로 형변환
vol = 1 / 3 * 3.14 * rad ** 2 * hei
suf = 3.14 * rad ** 2 + 3.14 * rad * sla
print("원뿔의 부피는", str(vol) + "입니다.", sep=": ")
print("원뿔의 겉넓이는", str(suf) + "입니다.")
```
