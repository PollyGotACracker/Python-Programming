### 텍스트 데이터 인사이트 분석기

**데이터 전처리(데이터 변환 및 불용어 필터링)**

```python
# 특수문자 목록
PUNCTUATIONS = ".,!?\"':;()-"

def clean_text(fp):
  text = ""

  for line in fp:
    ## 소문자 변환
    line = line.lower()

    ## 특수문자 제거
    for char in PUNCTUATIONS:
      line = line.replace(char, "")

    text += " " + line

  return text
```

**데이터 분석(단어 카운트)**

```python
# 불용어 집합
STOPWORDS = {"a", "the", "is", "of", "in", "to", "and", "i", "it", "that", \
             "this", "my", "was", "for", "as", "but", "so", "be", "he", \
             "me", "his", "him", "she", "her", "you", "your", "we", "our", "us", \
             "on", "were", "been", "are"}

def count_words(text):
  words = text.split()
  counts = {}

  for word in words:
    ## 문자열 공백 제거
    word.strip()

    ## 불용어일 경우 현재 반복 건너뛰기
    if word in STOPWORDS:
      continue

    # if word in counts:
    #   counts[word] += 1
    # else:
    #   counts[word] = 1
    counts[word] = counts.get(word, 0) + 1

  return counts
```

**데이터 시각화(텍스트 시각화)**

```python
## value-key 형태의 튜플을 리스트로 변환하고 정렬하는 함수
def get_sorted_list(counts_dict):
  temp_list = []

  for word in counts_dict:
    data_pair = (counts_dict[word], word)
    temp_list.append(data_pair)

  temp_list.sort(reverse=True)

  return temp_list

## 결과물을 그래프화하고 저장하는 함수
def print_save_graph(filename, sorted_list, tops = 10):
  re_fp = open(filename, "w")
  title = "== 데이터 분석 결과 보고서 =="
  print(title)

  re_fp.write(title + "\n")

  rank = 1

  for item in sorted_list:
    ## 지정한 순위권에서 벗어난 경우 반복 중지
    if rank > tops:
      break

    count = item[0] ## 출현횟수
    word = item[1] ## 키워드

    graph = "#" * (count // 100) ## // : 나눗셈의 몫을 반환
    result_line = format(rank, ">3n") + "." + format(word, "<15s") + ": " + \
                  graph + "(" + str(count) + ")\n"
    print(result_line)

    re_fp.write(result_line)

    rank += 1

  re_fp.close()
```

**결과 파일 저장 및 파일 닫기**

```python
##CWD, 파일목록 출력하고 파일이름 입력 받기
import os

print("현재 작업 디렉토리: ", os.getcwd())
print("현재 디렉토리 파일 목록: \n")
print(os.listdir(os.getcwd()))

##파일_변수 생성
filename = input("분석할 파일이름을 입력하세요.")

# print(filename)

da_fp = open(filename, "r")
text = clean_text(da_fp)
# print(text)

w_dict = count_words(text)
# print(w_dict)

count_list = get_sorted_list(w_dict)
# print(count_list)

print_save_graph("result.txt", count_list, 200)

da_fp.close()
```
