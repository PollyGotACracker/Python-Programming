## 3강 2절 - 연산자

```python
# 2 ** 10, 5 / 2 vs 5 // 2, 25 % 7, 8.4 % 0.9, number % 2

print(2 ** 10) # 거듭제곱
print(5 / 2)
print(5 // 2) # 몫
print(25 % 7) # 나머지(모듈러 연산)
print(8.4 % 0.9) # 모듈러 연산은 정수를 대상으로 하는 것이 좋다(부동소수점 오차)
```

```python
# max(2, 3, 4), min(2, 3, 4), round(3.141592, 2), round(3.415), abs(-3), pow(2, 3)
print(max(2, 3, 4))
print(min(2, 3, 4))
print(round(3.141592, 2))
print(round(3.415))
print(abs(-3))
print(pow(2, 3))
```
